```python
import pandas as pd
import matplotlib.pyplot as plt

# this is the code to plot the theoretical nine filters:



# the terminal velocity is supposed to be -2.7848 

# C = 2 m * g / (vterm) ^ 2 p * a

# this is the one with nine filters
 
def eulersMethod(y, t):

    m = 0.0088 - (0.0088 / 10)  
    A = 142.5 / 10000
    C = 0.853

    # C = 2mg / (Vterm)^2 * density * A

    density = 1.204

    g = -9.81 

    delta_t = 0.0001

    y = 1.36

    Vy = 0 

    y_lst = []

    t_lst = []


    while y > 0:

        a = g + ((C * density * A) / 2 * m) * Vy**2

        Vy += a * delta_t 

        y += Vy * delta_t

        t += delta_t 

        y_lst.append(y)

        t_lst.append(t)

    
    return y_lst, t_lst


y = 0

t = 0

y, t = eulersMethod(y, t)


plt.plot(t, y, label = 'Position with respect to time', c = 'blue')

plt.xlabel('Time')

plt.ylabel('Position')


# code to plot the experimental nine filters:




```
