```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# this is the code to plot the theoretical nine filters:



# the terminal velocity is supposed to be -2.7848 

# C = 2 m * g / (vterm) ^ 2 p * a

# this is the one with nine filters
 
def eulersMethod(y, t):

    m = 0.0088 - (0.0088 / 10)  
    A = 142.5 / 10000
    C = 0.853

    # C = 2mg / (Vterm)^2 * density * A

    density = 1.204

    g = -9.81 

    delta_t = 0.0001

    y = 1.36

    Vy = 0 

    y_lst = []

    t_lst = []


    while y > 0:

        a = g + ((C * density * A) / 2 * m) * Vy**2

        Vy += a * delta_t 

        y += Vy * delta_t

        t += delta_t 

        y_lst.append(y)

        t_lst.append(t)

    
    return y_lst, t_lst


y = 0

t = 0

y, t = eulersMethod(y, t)


plt.plot(t, y, label = 'Position with respect to time', c = 'blue')

plt.xlabel('Time')

plt.ylabel('Position')


# code to plot the experimental nine filters:

df = pd.read_csv("clean_data_many_filter.csv")
lst = []

for x in range(10):

    lst.append(df.iloc[0,10+x])

print(sum(lst)/10)

[-1.05576923,
-1.03461538,
-1.13115385,  
-1.15730769, 
-1.11461538,  
-1.05423077,  
-1.01923077,  
-1.06884615,  
-1.05653846,  
-1.03846154] 

position = df.loc[:,"Position (m) Run #3"]
time = df.loc[:,"Time (s) Run #3"]

plt.plot(time,position)

plt.legend(['Simulation', 'Experiment'])

Velocity = [-1.05576923,
-1.03461538,
-1.13115385,  
-1.15730769, 
-1.11461538,  
-1.05423077,  
-1.01923077,  
-1.06884615,  
-1.05653846,  
-1.03846154] 
averageV = sum(Velocity)/10
averageV

```
